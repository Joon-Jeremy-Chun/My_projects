```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools

plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
```

## 불러오기

```python
df = pd.read_csv('c://My_data//17년 세계전기차주제 검색 상대적 추의.csv')
```

## 나짜 인덱스가 있는데 원하는 범주로 새로 지정하여 엑셀에 저장해보기

```python
df1=df.iloc[1:]
```

```python
df1.reset_index()
```

```python
search_ls=list(((np.array(df1['카테고리: 모든 카테고리'].tolist()))))
search_ls1= list(map(int,search_ls))
```

```python
search_ls1
```

## 년도를 지정하는 과정
- n = 일단 월을 지정한다.(순서대로)
- m = 그다음 매월 1월을 년도의 기준점으로 잡기위해 12로 나누웠을때 나머지가 1인경우만 마크한다
- l = 그 마크에따라 2004년부터 순서대로 기입한다

```python
n = []
a = 0
for i in range(len(search_ls)) :
    a = a + 1
    n.append(a)
```

```python
m = []
for i in n :
    if i % 12 == 1 :
        m.append(1)
    else:
        m.append(0)
```

```python
l = []
a = 2003
for i in m :
    if i == 1 :
        a = a + 1
        l.append(a)
    else :
        l.append('')
```

## 데이터베이스의 값(list = search_ls) 와 지정한 년도 (list =l) 
을 가지고 그래프를 그려본다.

```python
plt.figure(figsize=(20,7))
plt.ylim((20,100))
plt.xlim([0,213])
plt.title('2004년~ 현재 세계 전기차주제 검색 변화',size =25)
plt.xlabel('기간(년)', labelpad = 10, size =20, weight = 800, )
plt.ylabel('상대적 검색비율', labelpad = 15 , size =20)
y = np.arange(len(search_ls1))
plt.bar(y, search_ls1,color = '#17becf')
plt.xticks(y, l, fontsize=15)


plt.show()
```

## 이제 평균 이동선을 그려본다

```python
# 5달 평균
m5 = []
a1=0
a2=0
a3=0
a4=0

for i in search_ls1 :
    a5 = i
    a4 = a5
    a3 = a4
    a2 = a3
    a1 = a2
    b = (a5+a4+a3+a3+a2+a1)/5
    m5.append(b)
```

```python
plt.figure(figsize=(20,7))
plt.ylim((0,100))
plt.xlim([0,213])
plt.title('2004년~ 현재 세계 전기차 관심도 변화',size =25)
plt.xlabel('기간(년)', labelpad = 10, size =20, weight = 800, )
plt.ylabel('상대적 검색비율', labelpad = 15 , size =20)
y = np.arange(len(search_ls1))
plt.plot(y, m5, color = '#17becf')
plt.xticks(y, l, fontsize=15)


plt.show()
```
