```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# 1년 대한민국 랭킹 18위
DF1 = pd.read_excel('c://My_data//electric_vehicle_serach_ranktoday%201-y.xlsx')

DF1.set_index('Country Name', inplace = True)


DF1['Search ratio'].sort_values().plot(kind = 'barh', grid = True, figsize = (7,20), color ='dodgerblue', width = 0.5)
plt.show()
```

```python
# 5년 대한민국 랭킹 23위
DF2 = pd.read_excel('c://My_data//electric_vehicle_serach_ranktoday%205-y.xlsx')

DF2.set_index('Country Name', inplace = True)

DF2['Search ratio'].sort_values().plot(kind = 'barh', grid = True, figsize = (7,20))
plt.show()
```
