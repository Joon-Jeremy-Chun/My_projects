### 1) Define dataframe - single company
-define empty array

-add on the data

-define DataFrame

```python
eps_his_msft_array = []

company_eps = si.get_earnings_history('MSFT')

for i in range(len(company_eps)):

      print('eps(actual)   :' , company_eps[i]['epsactual'])
      print('eps(estimate) :' , company_eps[i]['epsestimate'])
      print('date :', company_eps[i]['startdatetime'],company_eps[i]['timeZoneShortName'])
      print('\n')

      a = company_eps[i]['startdatetime']
      b = company_eps[i]['epsactual']
      c = company_eps[i]['epsestimate']
      X = [a,b,c]
      eps_his_msft_array.append(X)
MSFT_eps_his = pd.DataFrame(eps_his_msft_array)
MSFT_eps_his.columns = ['date','actual_eps','estimate_eps']
```
