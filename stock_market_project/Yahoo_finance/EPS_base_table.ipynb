##IV.Checking multiple companies by using 'for' statment

```python
#Semiconductor SOXX ETF holdings (top 10)
soxx_etf_holdings = ['INTC','AVGO','TXN','NVDA','QCOM','AMAT','MU','LRCX','KLAC','AMD'] #'TSMC (ADR) fail to recall 
```

```python
soxx_etf_holdings
```

### 1) To get the current PE ratio by 'for' statement

```python
for i in soxx_etf_holdings:
 print( i ,si.get_stats_valuation(i))
```

### 2) To get the date of coming earnings for multiple companies
- get_next_earnings_date()

```python
for i in soxx_etf_holdings:
 print( i ,si.get_next_earnings_date(i))
```

### 3)To get the date of earnings by 'for' satement
- get_earnings()

```python
for i in soxx_etf_holdings:
  print( i, si.get_earnings(i))
```

### 4) To get earnings history for mulitple companies

```python
X = ['INTC','AVGO','TXN','NVDA','QCOM','AMAT','MU','LRCX','KLAC','AMD']

for j in X:
  print ('---company abbreviation : ', j,'---')
  campany_eps = si.get_earnings_history(f'{j}')
  for i in range(len(j)):
      print('eps(actual)   :' , campany_eps[i]['epsactual'])
      print('eps(estimate) :' , campany_eps[i]['epsestimate'])
      print('date :', campany_eps[i]['startdatetime'],campany_eps[i]['timeZoneShortName'])
      print('\n')
```
