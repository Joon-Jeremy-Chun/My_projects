```python
test_Co = ['AVGO','AMD','NVDA','INTC','QCOM','TXN','ADI','KLAC','MCHP','NXPI','LRCX','AMAT','MRVL','MU','TSM']    

for j in test_Co:
  
    eps_his_array = []
    sleep(random.uniform(1,2))
    company_eps = si.get_earnings_history(j)
    print('------company abbreviation : ', j,'------')
    print('total data N : ' , len(company_eps))

    for i in range(len(company_eps)):


          # print('eps(actual)   :' , company_eps[i]['epsactual'])
          # print('eps(estimate) :' , company_eps[i]['epsestimate'])
          # print('date :', company_eps[i]['startdatetime'],company_eps[i]['timeZoneShortName'])
          # print('\n')

          a = company_eps[i]['startdatetime']
          b = company_eps[i]['epsactual']
          c = company_eps[i]['epsestimate']
          X = [a,b,c]
          eps_his_array.append(X)

    globals()[f'{j}_eps_his'] = pd.DataFrame(eps_his_array)
    globals()[f'{j}_eps_his'].columns = ['date','actual_eps','estimate_eps']
    print('complete creating dataframe: ', f'{j}_eps_his')
    print('\n')
```

### 3) storing the data

```python
test_Co = ['AVGO','AMD','NVDA','INTC','QCOM','TXN','ADI','KLAC','MCHP','NXPI','LRCX','AMAT','MRVL','MU','TSM']  

for j in test_Co:
  
    eps_his_array = []
    sleep(random.uniform(1,2))
    company_eps = si.get_earnings_history(j)
    print('------company abbreviation : ', j,'------')
    print('total data N : ' , len(company_eps))

    for i in range(len(company_eps)):
      
        try:


          # print('eps(actual)   :' , company_eps[i]['epsactual'])
          # print('eps(estimate) :' , company_eps[i]['epsestimate'])
          # print('date :', company_eps[i]['startdatetime'],company_eps[i]['timeZoneShortName'])
          # print('\n')

          a = company_eps[i]['startdatetime']
          b = company_eps[i]['epsactual']
          c = company_eps[i]['epsestimate']
          X = [a,b,c]
          eps_his_array.append(X)

        except IndexError as e:
          print(f'{j}', 'error')

    globals()[f'{j}_eps_his'] = pd.DataFrame(eps_his_array)
    globals()[f'{j}_eps_his'].columns = ['date','actual_eps','estimate_eps']
    globals()[f'{j}_eps_his'].to_csv(f'/content/drive/MyDrive/Colab Notebooks/datasets/stock_market/companies_eps_his/{j}_eps_his.csv')
    print('complete creating dataframe: ', f'{j}_eps_his')
    print('storning dataframe in (csv): ', f'{j}_eps_his')
    print('\n')
```
